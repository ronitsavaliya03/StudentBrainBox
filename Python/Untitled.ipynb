## Practical 1: Student Marks Analyzer
### Problem Definition
Store student data as a list of tuples. Remove duplicate marks, calculate average marks
and store result in a dictionary.
### Sample Input
students = [
(1, "Amit", [78, 85, 90, 85]),
(2, "Neha", [88, 92, 79, 88]),
(3, "Ravi", [70, 75, 80, 75])
]
### Sample Output
{1: ('Amit', 84.33), 2: ('Neha', 86.33), 3: ('Ravi', 75.0)}

## Practical 2: Inventory Management
### Problem Definition
Remove out-of-stock items and calculate total inventory value.
### Sample Input
inventory = {
 'pen': (10, 50), In [4]:
 'book': (100, 0),
 'pencil': (5, 100)
}
### Sample Output
{'pen': (10, 50), 'pencil': (5, 100)}
Total Value: 1000
